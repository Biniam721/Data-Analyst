```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_da = df[df['job_title_short'] == 'Data Analyst'].copy()
```

```python
df_da['job_posted_month_no'] = df_da['job_posted_date'].dt.month
df_da_explode = df_da.explode('job_skills')
df_da_explode = df_da_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
df_da_explode.loc['Total'] = df_da_explode.sum()
df_da_explode = df_da_explode[df_da_explode.loc['Total'].sort_values(ascending=False).index]
df_da_explode = df_da_explode.drop('Total')
df_da_explode 
```

```python
df_da_explode = df_da_explode.reset_index()
```

```python
df_da_explode = df_da_explode.reset_index()
df_da_explode['job_posted_month'] = df_da_explode['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_da_explode = df_da_explode.set_index('job_posted_month')
df_da_explode = df_da_explode.drop(columns='job_posted_month_no')

df_da_explode.iloc[:, :5].plot(kind='line')
plt.ylabel('Count')
plt.title('Top 5 Skills for Data Analysts per Month')
plt.xlabel('')
plt.show()
```
