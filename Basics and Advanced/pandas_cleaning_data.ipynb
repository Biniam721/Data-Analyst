```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[:10, 'salary_year_avg':'salary_hour_avg']
```

```python
salary_year = df['salary_year_avg'].median()
```

```python
salary_hour = df['salary_hour_avg'].median()
```

```python
df_filied = df

df_filied['salary_year_avg'] = df_filied['salary_year_avg'].fillna(salary_year)
df_filied['salary_hour_avg'] = df_filied['salary_hour_avg'].fillna(salary_hour)
```

```python
df_filied.loc[:10, 'salary_year_avg':'salary_hour_avg']
```

```python
df_unique = df_filied

df_unique = df_unique.drop_duplicates()
```
