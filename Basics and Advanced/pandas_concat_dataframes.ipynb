```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
df
```

```python
months = df['job_posted_month'].unique()
dict_months = {month: df[df['job_posted_month'] == month] for month in months}
dict_months['Jan']
```

Boss needs 2023 quarter 1

```python
quarter_1 = pd.concat([dict_months['Jan'], dict_months['Feb'], dict_months['Mar']], ignore_index=True)
```

```python
quarter_1['job_posted_month'].value_counts().plot(kind='bar')
```

```python
dict_months_pivot = quarter_1.pivot_table(index='job_title_short', columns='job_posted_month', aggfunc='size')
```

```python
dict_months_pivot.plot(kind='bar')
```
